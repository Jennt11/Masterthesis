```python
def plotting_rois(UMAP_roi_df, name,path):
    %matplotlib notebook
    sns.set_context("paper")
    sns.set_style("darkgrid")
    #sns.set(font_scale = 1.5)
    plt.figure(figsize=(10, 8))
    keys = ['OFC1', 'OFC2', 'OFC3', 'OFC4', 'OFC5', 'OFC6', 'OFC7', 'FP2', '25', 's24', 's32', '33', 'p24ab', 'p24c', 'p32',  'Putamen', 'NAcc', 'Caudate']
    sp = sns.scatterplot(x="x", y="y", hue='ROI', hue_order=keys, data=UMAP_roi_df, palette = colors18, legend = False, s=35, alpha=0.5)
    #sp = sns.scatterplot(x="x", y="y", hue='ROI', hue_order=keys, data=UMAP_roi_df, palette = colors18, legend = 'full', s=35, alpha=0.5)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., markerscale=2)
    plt.title(name, fontsize=26)
    plt.gca().invert_xaxis()
    #plt.setp(sp.get_legend().get_texts(), fontsize='24') # for legend text
    #plt.setp(sp.get_legend().get_title(), fontsize='32') # for the legend title
    plt.setp(sp,xticks=[-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10], yticks=[-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10])
    plt.xlabel('x', fontsize=24)
    plt.ylabel('y', fontsize=24)
    plt.tick_params(labelsize=22)
    sns.despine(top=True,offset=5)
    plt.tight_layout()
    plt.savefig(path, dpi=250)
```
