```python
# selfmade for 18 colors - 18 ROIs
colors18 = ['#ED1D0C','#808000', '#FA8072', '#7DDCAC', '#097E02', '#7988DF', '#38E1F5', '#1C0BEA', '#39374D', '#EC7BD9', '#560449', '#D192B8', '#F57A64', '#921701', '#A877EE', '#E2F337', '#BD11A1', '#89AB94']
```

```python
def plotting_rois(UMAP_roi_df, name,path):
    %matplotlib notebook
    sns.set()
    plt.figure(figsize=(10, 8))
    keys = ['OFC1', 'OFC2', 'OFC3', 'OFC4', 'OFC5', 'OFC6', 'OFC7', 'FP2', '25', 's24', 's32', '33', 'p24ab', 'p24c', 'p32',  'Putamen', 'NAcc', 'Caudate']
    sp = sns.scatterplot(x="x", y="y", hue='ROI', hue_order=keys, data=UMAP_roi_df, palette = colors18, legend = 'full', s=10, alpha=0.3)
    plt.legend(bbox_to_anchor=(0.9, 1), loc=2, borderaxespad=0.)
    plt.title(name, fontsize=18)
    plt.gca().invert_xaxis()
    plt.setp(sp,xticks=[-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10], yticks=[-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10])
    plt.xlabel('x', fontsize=14)
    plt.ylabel('y', fontsize=14)
    plt.savefig(path, dpi=250)
```

```python
#%matplotlib inline
plotting_rois(b_k, 'Baseline (Ketamine)', 'C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Plots/Baseline_Ket')
```
