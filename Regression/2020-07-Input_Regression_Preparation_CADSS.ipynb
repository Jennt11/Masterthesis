- for each participant you have one matrix 18 x 25 -> split in 18 x 10 (5 voxel from s24 + 5 voxel from s32) & flatten to be 1 x 180

```python
# list of voxels IDs for FC feature extraction
output = []
# function to flatten the list
def remove_nesting(input_list): 
 

    for i in input_list: 
        if type(i) == list: 
            remove_nesting(i) 
        else: 
            output.append(i) 
            
```

```python
# create feature names
fc_labels=[]
for ind in fc_feat.index:
    for col in output:
        name=str(col)+'_'+ind
        fc_labels.append(copy.deepcopy(name))
```

```python
# create features
def create_X(subj_list):
    
    # choose path according to saving directory dependend on treatment
    path = 'C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Correlation_matrices/ketFC_features-'
    
    #print(path)
    i=0
    
    for subj in subj_list:
        
        subj_name= str(subj)
        subj_path = path+subj_name
        # load features
        fc_feature = np.load(subj_path, allow_pickle=True)
        
        # extracting the columns for each input
        # X_1:
        X_feat= fc_feature[output].to_numpy()
        
        # flatten matrix from each particpant & stacking them up for all participants
        if i==0:
            # first round 
            X_1 = copy.deepcopy(X_feat.flatten())
        else:
            # all other rounds
            # vstack produces copy
            X_1 = np.vstack((X_1, X_feat.flatten()))

            
        i+=1 
    
    X_saving= pd.DataFrame(X_1)
    X_saving.columns=[fc_labels]
    X_saving.to_csv('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/X_CADSS.csv')
    np.save('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/X_CADSS.npy', X_1)
        
    return X_1
```
