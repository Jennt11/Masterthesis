# Reading in ROI labels and concatenating them

```python
import numpy as np
import pandas as pd
import pickle
```

```python
labels_path_pfc = 'C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Labeling/PFC_ROIs.csv'
labels_path_str = 'C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Labeling/Striatum_ROIs.csv'
labels_pfc = pd.read_csv(labels_path_pfc, engine='c', header =None)
#data = pd.read_csv(dat_path, engine='c').astype('float32') 
labels_str = pd.read_csv(labels_path_str, engine='c', header=None)
```

```python
mask_label_pfc_new= pd.read_csv('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Labeling/Mask_ind_pfc_new.csv', engine='c', header = None)
mask_label_str_new= pd.read_csv('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Labeling/Mask_ind_str_new.csv', engine='c', header = None)
```

```python
mask_label_pfc_new
```

```python
mask_ind= pd.concat([mask_label_pfc_new, mask_label_str_new], axis=0, ignore_index=True, sort=False)
mask_ind.columns=['mask_id']
```

```python
mask_ind
```

```python
roi_labels = pd.concat([labels_pfc, labels_str], axis=0, ignore_index=True, sort=False)
roi_labels.columns = ['label']
```

```python
roi_labels
```

```python
label_ID= pd.concat([mask_ind, roi_labels], axis=1)
label_ID
```

```python
# saving them with the name roi_labels
roi_labels.to_pickle('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Labeling/roi_labels')
mask_ind.to_pickle('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Labeling/mask_ind')
label_ID.to_pickle('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Labeling/labels_ID')
```
