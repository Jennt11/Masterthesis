**Ketamine group**

```python
# first matrix of zeros
A = np.zeros((11883,11883))
for subj in subj_ket_new:
    
    file_name = 'corr_mat_all_ses-1h_sub-' + subj + '.csv'
    # path of the current subject
    curr_subj_path = os.path.join(path_ket, file_name)
    # read in current subject as dataframe
    curr_subj_df = pd.read_csv(curr_subj_path, header=None)
    # change dataframe to numpy array
    curr_sub = curr_subj_df.to_numpy()
    #print('Row 665: ', curr_sub[665])
    # z-transform
    z_curr_sub = np.arctanh(curr_sub)
    # adding all matrices
    mean_mat = np.add(z_curr_sub, A)
    print(subj, ': ' , mean_mat[665])
    # making proper copy of mean_mat
    A = mean_mat.copy()

#np.save('C:/Users/canlab/Desktop/Masterthesis_python/ket/Ketamine_z_FC_added', mean_mat)
# calculating mean of all matrices
z_mean_mat_ket = mean_mat / len(subj_ket_new)  
np.save('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Ket/Ketamine_z_mean_FC', z_mean_mat_ket)
```

**Placebo group**

```python
# first matrix of zeros
A = np.zeros((11883,11883))
for subj in subj_placebo_new:
    
    file_name = 'corr_mat_all_ses-1h_sub-' + subj + '.csv'
    # path of the current subject
    curr_subj_path = os.path.join(path_placebo, file_name)
    # read in current subject as dataframe
    curr_subj_df = pd.read_csv(curr_subj_path, header=None)
    # change dataframe to numpy array
    curr_sub = curr_subj_df.to_numpy()
    # Fisher transform
    z_curr_sub = np.arctanh(curr_sub)
    # adding
    mean_mat = np.add(z_curr_sub, A)
    
    print(subj, ': ' , mean_mat)
    # making proper copy of mean_mat
    A = mean_mat.copy()
    
# saving matrices    
#np.save('C:/Users/canlab/Desktop/Masterthesis_python/placebo/Placebo_z_FC_added', mean_mat)
mean_mat_placebo = mean_mat / len(subj_placebo_new) 
np.save('C:/Users/canlab/Desktop/Masterthesis_python/ReDoneAnalysis20200420/Placebo/Placebo_z_mean_FC', mean_mat_placebo)
```
