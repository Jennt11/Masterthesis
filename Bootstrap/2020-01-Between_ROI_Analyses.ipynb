#### Only look at distances from all voxels to striatum centroids

```python
baseline = voxels_to_centroids_b[['Putamen','NAcc', 'Caudate', 'ROI_voxel']]
ketamine = voxels_to_centroids_k[['Putamen','NAcc', 'Caudate', 'ROI_voxel']]
placebo = voxels_to_centroids_p[['Putamen','NAcc', 'Caudate', 'ROI_voxel']]
```

```python
baseline
```

```python
str_25_b = baseline[baseline['ROI_voxel']=='25']
str_25_k = ketamine[baseline['ROI_voxel']=='25']
str_25_p = placebo[baseline['ROI_voxel']=='25']
```

### ROI 25

Scatter plot: distance of voxels of area 25 (128) to all striatum ROIs

Putamen

```python
putamen = pd.concat([str_25_b[['Putamen']],str_25_k[['Putamen']],str_25_p[['Putamen']]], axis=1)
putamen.columns = ['Baseline', 'Ketamine', 'Placebo']
putamen = putamen.reset_index(drop=True)
putamen
```

```python
%matplotlib notebook
plt.figure(figsize=(8,5))
sp = sns.scatterplot(hue=['Baseline', 'Ketamine', 'Placebo'], data=putamen, legend = 'full', s=10)
plt.legend(bbox_to_anchor=(0.9, 1), loc=2, borderaxespad=0.)
plt.title('Distance voxels 25 to putamen', fontsize=18)
plt.xlabel('Voxel', fontsize=14)
plt.ylabel('Distance', fontsize=14)
plt.savefig('/Users/jenniferthem/Masterthesis_Data/January2020/dist_25_putamen.png', dpi=250)
```

```python
%matplotlib notebook
keys = ['Baseline', 'Ketamine', 'Placebo']
# Iterate through the five airlines
for curr_cond in keys:
    # Subset to the airline
    subset = putamen[curr_cond]

    sns.distplot(subset, hist = False, kde = True,
                 kde_kws = {'linewidth': 1},
                 label = curr_cond)
    
# Plot formatting
#fig.set_size_inches(18.5, 10.5)
#fig.subplots(figsize=(20,15))
plt.legend(prop={'size': 8}, title = 'Condition')
plt.title('25-Putamen')
plt.xlabel('Distance')
plt.ylabel('Density')
#plt.show()
plt.savefig('/Users/jenniferthem/Masterthesis_Data/January2020/Density_25_putamen.png', dpi=250)
```
