```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io
import scipy
import seaborn as sns
import tikzplotlib as tpl
import pickle
```

Load data: df with distance of each voxel to their respective ROI centroid

```python
b = np.load('/Users/jenniferthem/Masterthesis_Data/January2020/b_df_labeled.npy', allow_pickle=True)
k = np.load('/Users/jenniferthem/Masterthesis_Data/January2020/k_df_labeled.npy', allow_pickle=True)
p = np.load('/Users/jenniferthem/Masterthesis_Data/January2020/p_df_labeled.npy', allow_pickle=True)
```

Transform to dataframe for plotting

```python
b_df_labeled_ = pd.DataFrame(b)
k_df_labeled_ = pd.DataFrame(k)
p_df_labeled_ = pd.DataFrame(p)
```

```python
b_df_labeled_.columns= ['Voxel', 'Distance', 'ROI']
k_df_labeled_.columns= ['Voxel', 'Distance', 'ROI']
p_df_labeled_.columns= ['Voxel', 'Distance', 'ROI']
```

### Scatterplots

```python
%matplotlib notebook
plot = sns.scatterplot(x ='Voxel', y= 'Distance', hue='ROI', data=b_df_labeled_, palette = colors18, legend = 'full', s=10)
plt.legend(bbox_to_anchor=(0.9, 1), loc=2, borderaxespad=0.)
plt.title('Baseline', fontsize=18)
plt.savefig('/Users/jenniferthem/Masterthesis_Data/January2020/baseline_distance_centroid.png', dpi=250)
#tpl.save("baseline_distance_centroid.tex")
```

```python
%matplotlib notebook
plot = sns.scatterplot(x ='Voxel', y= 'Distance', hue='ROI', data=k_df_labeled_, palette = colors18, legend = 'full', s=10)
plt.legend(bbox_to_anchor=(0.9, 1), loc=2, borderaxespad=0.)
plt.title('Ketamine', fontsize=18)
plt.savefig('/Users/jenniferthem/Masterthesis_Data/January2020/Ketamine_distance_centroid.png', dpi=250)
```
