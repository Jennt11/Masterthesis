## Function for calculating centroids

```python
# input: dataframe containing x,y (UMAP),ROI --> 3D
def calculate_centroid(input_df):
    # split dataframe according to ROIs
    split_df = input_df.groupby('ROI')
    # arithmetic mean values for all ROIs used as centroids of the ROIs
    ROI_mean = split_df.mean()
    #display(ROI_mean)
    # mean df to dictionary
    ROI_mean_dict = ROI_mean.to_dict()
    # save all distances
    centroid_dist = []
    # dictionary to save distances to centroid for each ROI
    dist_to_centroid = {}
    keys = ['OFC1', 'OFC2', 'OFC3', 'OFC4', 'OFC5', 'OFC6', 'OFC7', 's32', 'FP2', '25', 's24', '33', 'p32', 'p24ab', 'p24c', 'Putamen', 'NAcc', 'Caudate']
    # loop through ROI_mean_dict to get the mean for the ROI
    for entry in keys:
        dist =0
        # get mean values for ROIs
        x_mean = ROI_mean_dict['x'][entry]
        y_mean = ROI_mean_dict['y'][entry]
        
        for i in range(18):
            if entry==list(split_df)[i][0]:
                # set ROI_df to this i for specific ROI
                ROI_df = list(split_df)[i]
                break
        # extract all x and y values for each ROI (-> ROI label is 3rd column)
        ROI_coord = ROI_df[1].iloc[:,0:2].to_numpy()
        
        for j in range(len(ROI_coord)):
            x = ROI_coord[j][0]
            y = ROI_coord[j][1]
            dist += np.sqrt(((x-x_mean)**2)+((y-y_mean)**2))   
        # save distance to centroid in dictionary with ROI
        dist_to_centroid[entry] = dist/len(ROI_coord)

    return dist_to_centroid
```

```python

# input: dataframe containing x,y (UMAP),ROI --> 3D
def calculate_centroid(input_df):
    # split dataframe according to ROIs
    split_df = input_df.groupby('ROI')
    # arithmetic mean values for all ROIs used as centroids of the ROIs
    ROI_mean = split_df.mean()
    #display(ROI_mean)
    # mean df to numpy
    ROI_mean_arr = ROI_mean.to_numpy()
    # save all distances
    centroid_dist = []
    for i in range(18):
        ROI_df = list(split_df)[i]
        # dataframe of one ROI
        # extracting the x and y columns and making it numpy array
        ROI_coord = ROI_df[1].iloc[:,0:2].to_numpy()
        # centroids (x,y) of this ROI
        curr_mean = ROI_mean_arr[i]
        dist=0
        # calculating distance centroid to all other points
        for j in range(len(ROI_coord)):
            x = ROI_coord[j][0]
            y = ROI_coord[j][1]
            dist += np.sqrt(((x-curr_mean[0])**2)+((y-curr_mean[1])**2))
        
        centroid_dist.append(dist/len(ROI_coord))
    centroid_dist_df = pd.DataFrame(centroid_dist)
    centroid_dist_df.columns=['centroid distance']
    #mean_centroid_df = pd.concat(ROI_mean, centroid_dist_df)
    return centroid_dist, centroid_dist_df, ROI_mean, ROI_mean_arr
```
