## Function for calculating centroids

Calculating centroids + averaged distance of voxels within ROI

```python
# input: dataframe containing x,y (UMAP),ROI --> 3D
def calculate_centroid(input_df):
    # split dataframe according to ROIs
    split_df = input_df.groupby('ROI')
    # arithmetic mean values for all ROIs used as centroids of the ROIs
    ROI_mean = split_df.mean()
    #display(ROI_mean)
    # mean df to dictionary
    ROI_mean_dict = ROI_mean.to_dict()
    # save all distances
    centroid_dist = []
    # dictionary to save distances to centroid for each ROI
    dist_to_centroid = {}
    keys = ['OFC1', 'OFC2', 'OFC3', 'OFC4', 'OFC5', 'OFC6', 'OFC7', 's32', 'FP2', '25', 's24', '33', 'p32', 'p24ab', 'p24c', 'Putamen', 'NAcc', 'Caudate']
    # loop through ROI_mean_dict to get the mean for the ROI
    for entry in keys:
        dist =0
        # get mean values for ROIs
        x_mean = ROI_mean_dict['x'][entry]
        y_mean = ROI_mean_dict['y'][entry]
        
        for i in range(18):
            if entry==list(split_df)[i][0]:
                # set ROI_df to this i for specific ROI
                ROI_df = list(split_df)[i]
                break
        # extract all x and y values for each ROI (-> ROI label is 3rd column)
        ROI_coord = ROI_df[1].iloc[:,0:2].to_numpy()
        
        for j in range(len(ROI_coord)):
            x = ROI_coord[j][0]
            y = ROI_coord[j][1]
            dist += np.sqrt(((x-x_mean)**2)+((y-y_mean)**2))   
        # save distance to centroid in dictionary with ROI
        dist_to_centroid[entry] = dist/len(ROI_coord)

    return dist_to_centroid
```

All voxels of one ROI to ROI centroid -  not averaged
returns a dictionary for each ROI centroid and the respective ROI voxels

```python
# input: dataframe containing x,y (UMAP),ROI --> 3D
def calculate_dist_voxel_to_resp_centroid(input_df):
    # split dataframe according to ROIs
    split_df = input_df.groupby('ROI')
    # arithmetic mean values for all ROIs used as centroids of the ROIs
    ROI_mean = split_df.mean()
    #display(ROI_mean)
    # mean df to dictionary
    ROI_mean_dict = ROI_mean.to_dict()
    # save all distances
    centroid_dist = []
    # dictionary to save distances to centroid for each ROI
    dist_to_centroid = {}
    keys = ['OFC1', 'OFC2', 'OFC3', 'OFC4', 'OFC5', 'OFC6', 'OFC7', 's32', 'FP2', '25', 's24', '33', 'p32', 'p24ab', 'p24c', 'Putamen', 'NAcc', 'Caudate']
    # loop through ROI_mean_dict to get the mean for the ROI
    k =0
    roi_label_list = []

    for entry in keys:
        dist =0
        # get mean values for ROIs
        x_mean = ROI_mean_dict['x'][entry]
        y_mean = ROI_mean_dict['y'][entry]
        
        for i in range(18):
            if entry==list(split_df)[i][0]:
                # set ROI_df to this i for specific ROI
                ROI_df = list(split_df)[i]
                break
        # extract all x and y values for each ROI (-> ROI label is 3rd column)
        ROI_coord = ROI_df[1].iloc[:,0:2].to_numpy()
        
        for j in range(len(ROI_coord)):
            x = ROI_coord[j][0]
            y = ROI_coord[j][1]
            dist = np.sqrt(((x-x_mean)**2)+((y-y_mean)**2))
            # new ROI label list to later have a ROI labeling for the voxels
            roi_label_list.append(entry)
            # dictionary that saves dist from each voxel to their respective ROI centroid
            dist_to_centroid[k] = dist
            k += 1
            
    np.save('/Users/jenniferthem/Masterthesis_Data/January2020/Roi_labeling_centroids', roi_label_list)
    return dist_to_centroid, roi_label_list
```

Distance of all voxels to ROI centroids

```python
def calculate_dist_voxels_to_centroids(input_df, embedding):
    # split dataframe according to ROIs
    split_df = input_df.groupby('ROI')
    # arithmetic mean values for all ROIs used as centroids of the ROIs
    ROI_mean = split_df.mean()
    #display(ROI_mean)
    # mean df to dictionary
    ROI_mean_dict = ROI_mean.to_dict()
    # save all distances
    centroid_dist = []
    # dictionary to save distances to centroid for each ROI
    dist_to_centroid = {}
    keys = ['OFC1', 'OFC2', 'OFC3', 'OFC4', 'OFC5', 'OFC6', 'OFC7', 's32', 'FP2', '25', 's24', '33', 'p32', 'p24ab', 'p24c', 'Putamen', 'NAcc', 'Caudate']
    # loop through ROI_mean_dict to get the mean for the ROI
    for entry in keys:
        dist =0
        # get mean values for ROIs
        x_mean = ROI_mean_dict['x'][entry]
        y_mean = ROI_mean_dict['y'][entry]
        roi_to_voxel = []
        for j in range(len(embedding)):
            x = embedding[j][0]
            y = embedding[j][1]
            dist = np.sqrt(((x-x_mean)**2)+((y-y_mean)**2))  
            roi_to_voxel.append(dist)
        # save distance to centroid in dictionary with ROI
        dist_to_centroid[entry] = roi_to_voxel

    return dist_to_centroid
```

Average distance of all voxels within one ROI to the ROI centroids

```python
# input: dataframe containing x,y (UMAP),ROI --> 3D
def calculate_dist_all_centroids(input_df):
    # split dataframe according to ROIs
    split_df = input_df.groupby('ROI')
    # arithmetic mean values for all ROIs used as centroids of the ROIs
    ROI_mean = split_df.mean()
    #display(ROI_mean)
    # mean df to dictionary
    ROI_mean_dict = ROI_mean.to_dict()
    # save all distances
    centroid_dist = []
    # dictionary to save distances to centroid for each ROI
    dist_to_centroid = {}
    keys = ['25', '33', 'Caudate', 'FP2', 'NAcc', 'OFC1', 'OFC2', 'OFC3', 'OFC4', 'OFC5', 'OFC6', 'OFC7', 'Putamen', 'p24ab', 'p24c','p32', 's24', 's32']
    # loop through ROI_mean_dict to get the mean for the ROI
    for entry in keys:
        # get mean values for ROIs
        x_mean = ROI_mean_dict['x'][entry]
        y_mean = ROI_mean_dict['y'][entry]
        # loop through all ROIs
        dict_roi_dist = {}
        for i in range(18):
            
            roi_name = list(split_df)[i][0]
            # set ROI_df to this i for specific ROI
            ROI_df = list(split_df)[i]
            # extract all x and y values for each ROI (-> ROI label is 3rd column)
            ROI_coord = ROI_df[1].iloc[:,0:2].to_numpy()
            # loop through all voxels of one ROI
            dist =0
            
            for j in range(len(ROI_coord)):
                x = ROI_coord[j][0]
                y = ROI_coord[j][1]
                dist += np.sqrt(((x-x_mean)**2)+((y-y_mean)**2)) 
                #print('J:', j)
            #print('ROI_name ', roi_name, ' i: ', i)
            # save distance to centroid in dictionary with ROI
            dict_roi_dist[roi_name] = dist/len(ROI_coord)
            #print(dict_roi_dist)
        name = entry + ' centroid' 
        # average distance of all voxels within one ROI to the ROI centroids
        dist_to_centroid[name] = dict_roi_dist
    return dist_to_centroid
```
