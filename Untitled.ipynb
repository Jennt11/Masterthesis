```python
import multiprocessing as mp
import pandas as pd
import numpy as np
from scipy.spatial import distance
```

```python
# Step 1: Init multiprocessing.Pool()
pool = mp.Pool(mp.cpu_count())
```

```python
data_path = '/Users/jenniferthem/Masterthesis_Data/z_m_corr_mat_baseline.csv'
corr_mat = pd.read_csv(data_path, engine='c', header = None).astype('float32') 
```

```python
pd.set_option('use_inf_as_na', True)
```

```python
corr_mat_df = corr_mat.fillna(1)
```

```python
display(corr_mat_df)
```

```python
FC_mat = corr_mat_df.to_numpy()
```

```python
FC_mat
```

```python
corr_vec = np.zeros((11883, 11883), dtype='float')
```

```python
corr_vec
```

```python
u= FC_mat[0,:]
u
```

```python

def calculate_corr_distance():
    i = 0;
    
    while i < 11883:
        j=0;
        while j < 11883:
            u= FC_mat[i,:]
            v= FC_mat[j,:]
            corr_vec[i,j] = distance.correlation(u,v)
            j = j+1
            
        i = i +1
            
    return corr_vec
```

```python
pool = mp.Pool(mp.cpu_count())
```

```python
results = [pool.apply(calculate_corr_distance())]
pool.close()  
```
